# This Notebook is the case study part of the thesis.

```python
import fbprophet
import statsmodels as sm
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import particles  # core module
from particles import distributions as dists  # where probability distributions are defined
from particles import state_space_models as ssm  # where state-space models are defined
from src.Smoother import ma_smoother, exp_smoother,ld_smoother, sqd_smoother, inv_smoother, read_all_data #self defined smoother clasees
```

```python
stock_data = read_all_data().query('Name=="MMM"')
ma_sm = ma_smoother(10,stock_data)
```

```python
ma_sm.get_smoothed_pred()
```

```python
#compare AA and variance between different methods in a grid set of window size
grid_h = list(range(5,100,5))
val_dict = {} #dict stores values
val_dict['ma'] = []
val_dict['ema'] = []
val_dict['ld'] = []
val_dict['sqd'] = []
val_dict['inv'] = []
for h in grid_h:
    tmp_model = ma_smoother(h, stock_data)
    val_dict['ma'].append((tmp_model.age(h),tmp_model.get_variance(h)))
    tmp_model = exp_smoother(h, stock_data, alpha=0.7)
    val_dict['ema'].append((tmp_model.age(h), tmp_model.get_variance(h)))
    tmp_model = ld_smoother(h, stock_data)
    val_dict['ld'].append((tmp_model.age(h),tmp_model.get_variance(h)))
    tmp_model = sqd_smoother(h, stock_data)
    val_dict['sqd'].append((tmp_model.age(h),tmp_model.get_variance(h)))
    tmp_model = inv_smoother(h, stock_data)
    val_dict['inv'].append((tmp_model.age(h),tmp_model.get_variance(h)))
    
# exp_sm = exp_smoother(10,stock_data, alpha=0.7)
```

```python
#exponential smoothing test
ema_val = []
grid_alpha = np.arange(0.1,0.9,0.1)
for alpha in grid_alpha:
    tmp_model = exp_smoother(50, stock_data, alpha)
    ema_val.append((tmp_model.age(50), tmp_model.get_variance(50)))
```

```python
#visualization
PROJECT_DIR = os.getcwd()
IMAGE_DIR = os.path.join(PROJECT_DIR, './images/')
plt.plot([x[0] for x in val_dict['ma']], [x[1] for x in val_dict['ma']], label='MA')
plt.legend()
plt.plot([x[0] for x in ema_val], [x[1] for x in ema_val], label='EMA')
plt.legend()
plt.plot([x[0] for x in val_dict['ld']], [x[1] for x in val_dict['ld']], label='LD')
plt.legend()
plt.plot([x[0] for x in val_dict['sqd']], [x[1] for x in val_dict['sqd']], label='SQD')
plt.legend()
plt.plot([x[0] for x in val_dict['inv']], [x[1] for x in val_dict['inv']], label='INV')
plt.legend()
plt.xlabel('AA')
plt.ylabel('Variance')
plt.savefig(IMAGE_DIR+'AA-VAR-Comp.jpg')
```
